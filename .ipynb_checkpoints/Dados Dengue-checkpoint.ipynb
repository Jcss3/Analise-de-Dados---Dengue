## Agrupando Todos os Dados

```python
def UniaoDataframes(dataFrame1,dataFrame2,dataFrame3,dataFrame4,dataFrame5,dataFrame6):
    
    #convertendo os tipos para forçar a união
    dataFrame1 = dataFrame1[dataFrame1.columns.tolist()].astype('category')
    dataFrame2 = dataFrame2[dataFrame2.columns.tolist()].astype('category')
    dataFrame3 = dataFrame3[dataFrame3.columns.tolist()].astype('category')
    dataFrame4 = dataFrame4[dataFrame4.columns.tolist()].astype('category')
    dataFrame5 = dataFrame5[dataFrame5.columns.tolist()].astype('category')
    dataFrame6 = dataFrame6[dataFrame6.columns.tolist()].astype('category')
    
    dados1 = pd.merge(dataFrame1,dataFrame2,how='outer')
    dados2 = pd.merge(dataFrame3,dataFrame4,how='outer')
    dados3 = pd.merge(dataFrame5,dataFrame6,how='outer')
    
    # Como gerou novos dataframes temos que refazer a conversão
    dados1 = dados1[dados1.columns.tolist()].astype('category')
    dados2 = dados2[dados2.columns.tolist()].astype('category')
    dados3 = dados3[dados3.columns.tolist()].astype('category')
    
    
    dados_parcial = pd.merge(dados1,dados2,how='outer')
    
    # Como gerou novos dataframes temos que refazer a conversão
    dados_parcial = dados_parcial[dados_parcial.columns.tolist()].astype('category')
    
    dados = pd.merge(dados_parcial,dados3,how='outer')
    return dados
```

```python
Dados = UniaoDataframes(casos2013,casos2014,casos2015,casos2016,casos2017,casos2018)
Dados
```

### Criando a nova Base de Dados

```python
Dados.to_csv('Dados_2013_2018.csv',index_label=False)
```
