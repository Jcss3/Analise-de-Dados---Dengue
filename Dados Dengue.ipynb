## Agrupando Todos os Dados

```python
def UniaoDataframes(dataFrame1,dataFrame2,dataFrame3,dataFrame4,dataFrame5,dataFrame6):
    dados1 = pd.merge(dataFrame1,dataFrame2,how='outer')
    dados2 = pd.merge(dataFrame3,dataFrame4,how='outer')
    dados3 = pd.merge(dataFrame5,dataFrame6,how='outer')
    dados_parcial = pd.merge(dados1,dataFrame2,how='outer')
    dados = pd.merge(dados_parcial,dados3,how='outer')
    return dados
```

```python
teste = UniaoDataframes(casos2013,casos2014,casos2015,casos2016,casos2017,casos2018)
```
